```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt

# Pre-processing
from sklearn import preprocessing

# Anomaly Detection
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
from sklearn.metrics import roc_auc_score
```

```python
data1=pd.read_csv('data/predictive-maintenance/predictive-maintenance.csv')
data1
```

```python
##Pre-processing
## UDI was dropped since it is basically a second index - it was unneeded. Allo the machine failure columns(TWF etc.)
## were dropped as according to specifications they were rolled into Machine failure. Product ID was also dropped as it is
## not a logical indicator of anomalies. This leaves us with 7 columns(from 14). Machine failure is our y axis.
data1.drop(['Product ID', 'UDI', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF'], axis='columns', inplace=True)
data1
```

```python
## Just doublechecking that there's only these three "Types"
print("Unique Types: ", data1['Type'].unique())
## then converting them into 0, 1, and 2 respectively because numbers are much easier to account for than letters
data1.loc[(data1.Type == 'M'), ('Type')] = 0
data1.loc[(data1.Type == 'L'), ('Type')] = 1
data1.loc[(data1.Type == 'H'), ('Type')] = 2

## It's easier to reuse algorithms from class that assume anomalies are -1 and non-anomalies are 1, so
## we'll do it like this - also the other database does the same
data1.loc[(data1['Machine failure']==1), ('Machine failure')] = -1
data1.loc[(data1['Machine failure']==0), ('Machine failure')] = 1
```

```python
X = data1.drop(['Machine failure'], axis='columns')
y = data1['Machine failure'].values
## Scaling each feature so they fit within a range on a training set (0-1 here) 
scaler = preprocessing.MinMaxScaler()
X_data = X.values

X_scaled = scaler.fit_transform(X_data)
X = X_scaled

auc_scores = []
for i in range(5, 51, 5):
    detector_lof = LocalOutlierFactor(n_neighbors=i, novelty=True)
    detector_lof.fit(X)
    y_score = detector_lof.decision_function(X)
    auc = roc_auc_score(y, y_score)
    auc_scores.append(auc)

# Plot the relationship between auc and k
plt.plot(range(5, 51, 5), auc_scores, marker='x')
plt.xlabel('K')
plt.ylabel('AUC')
plt.show()

print('\n Anomaly rate: %f \n' % (-1.0*(y-1).sum()/2.0/X.shape[0]))
```

As the graph demonstrates, a value of 35 k (i.e. n_neighbors=35) returns the highest AUC (and hence, the best accuracy rate), so we'll use that value to assess LOF vs other methods.

```python
detector_lof = LocalOutlierFactor(n_neighbors=35, novelty=True)
detector_lof.fit(X)
## Note - I'm printing all of the AUC scores after I've executed all three methods, so that they're easier to compare.

y_scorelof = detector_lof.decision_function(X)
auclof = roc_auc_score(y, y_scorelof)
y_predlof = detector_lof.predict(X)
auclof2 = roc_auc_score(y, y_predlof)
```

```python
auc_scores = []
sample_sizes = []
for i in range(3, 15):
    sample_sizes.append(2**i)
    detector_if = IsolationForest(max_samples=2**i)
    detector_if.fit(X)
    y_score = detector_if.decision_function(X)
    auc = roc_auc_score(y, y_score)
    auc_scores.append(auc)

# Plot the relationship between auc and sample size
plt.plot(sample_sizes, auc_scores, marker='x')
plt.xscale('log')
plt.xlabel('Sample size (log)')
plt.ylabel('AUC')
plt.show()
```

Again, the graph demonstrates a K value that we will use for our comparison between methods later. In this case, I chose to use 2^11, although I considered 2^5 as a valid choice as well, sacrificing some sample size for a small increase in accuracy

```python
detector_if = IsolationForest(max_samples=2**11)
detector_if.fit(X)
y_score = detector_if.decision_function(X)
## Note - I'm printing all of the AUC scores after I've executed all three methods, so that they're easier to compare.

aucif = roc_auc_score(y, y_score)
y_predif = detector_if.predict(X)
```

```python
# Create the Robust covariance model
ee = EllipticEnvelope()
ee.fit(X)
## Note - I'm printing all of the AUC scores after I've executed all three methods, so that they're easier to compare.

y_scoreee = ee.decision_function(X)
aucee = roc_auc_score(y, y_scoreee)
y_predee = ee.predict(X)
```

```python
print('\n AUC LOF score: %f\n' % auclof)
print('\n AUC LOF Label score: %f\n' % auclof2)
print('\n AUC IF score: %f\n' % aucif)
print('\n AUC IF Label score: %f\n' % roc_auc_score(y, y_predif))
print('\n AUC EE score: %f\n' % aucee)
print('\n AUC EE Label score: %f\n' % roc_auc_score(y, y_predee))
```

It can be seen that all three methods enjoy an substantial increase in accuracy over random chance, at the very least. The robust covariance model boasted the highest average score(for label and raw AUC) across all 3 models. A 17% improvement(at least) over random chance is valuable.
